```python
import numpy as np
from scipy.optimize import fminbound, fmin_bfgs
from scipy import interp

class IndividualProblem:

    def __init__(self, r=0.01, beta=0.96, Pi=((0.6, 0.4), (0.05, 0.95)),
                 z_vals=(0.5, 1), b=0, grid_max=16, grid_size=50,
                 u=np.log, du=lambda x: 1/x):
        self.u, self.du = u, du
        self.r, self.R = r, 1 + r
        self.beta, self.b = beta, b
        self.Pi, self.z_vals = np.array(Pi), tuple(z_vals)
        self.asset_grid = np.linspace(-b, grid_max, grid_size)

    def bellman_operator(self, V, return_policy=False):
        """
        Approx. Bellman operator returns updated value function TV
        (or the V-greedy policy c if return_policy is True).

        """
        # Shorten and define variables
        R, Pi, beta, u, b = self.R, self.Pi, self.beta, self.u, self.b
        asset_grid, z_vals = self.asset_grid, self.z_vals
        new_V = np.empty(V.shape)
        new_c = np.empty(V.shape)
        z_idx = list(range(len(z_vals)))

        # Linear interpolation of V along asset grid
        vf = lambda a, i_z: interp(a, asset_grid, V[:, i_z])

        # Solve RHS of Bellman equation
        for i_a, a in enumerate(asset_grid):
            for i_z, z in enumerate(z_vals):
                
                # Using fminbound - I managed to obtain the answer
                def obj(c):
                    y = sum(vf(R * a + z - c, j) * Pi[i_z, j] for j in z_idx)
                    return - u(c) - beta * y
                
                c_star = fminbound(obj, np.min(z_vals), R * a + z + b)
                
                # Using fmin_bfgs - I obtain the error output below
                def f(x):
                    # Obj is the Lagrangian with min and max search values as constraints
                    y = sum(vf(R * a + z - x[0], j) * Pi[i_z, j] for j in z_idx)
                    return - u(x[0]) - beta * y - x[1]*(np.min(z_vals)-x[0]) - x[2]*(x[0]-R*a-z-b)
                
                c_star_bfgs, price1, price2 = fmin_bfgs(f, (0.00001,0.00001,0.00001), disp=True)
                
                # The graph obtained at the bottom is the answer that I want to replicate
                new_c[i_a, i_z], new_V[i_a, i_z] = c_star, -obj(c_star)
                # as shown this uses c_star (fminbound) instead of c_star_bfgs
                # if c_star_bfgs is used instead, an empty histogram is obtained

        if return_policy:
            return new_c
        else:
            return new_V

    def initialize(self):
        # Shorten and define variables
        R, beta, u, b = self.R, self.beta, self.u, self.b
        asset_grid, z_vals = self.asset_grid, self.z_vals
        shape = len(asset_grid), len(z_vals)
        V = np.empty(shape)

        # Populate V
        for i_a, a in enumerate(asset_grid):
            for i_z, z in enumerate(z_vals):
                c_max = R * a + z + b
                V[i_a, i_z] = u(c_max) / (1 - beta)

        return V
```

```python
#==== This cell contains functions from additional libraries ====#
from numpy import cumsum
from numpy.random import uniform

class DiscreteRV(object):
    """
    Generates an array of draws from a discrete random variable with
    vector of probabilities given by q.

    Parameters
    ----------
    q : array_like(float)
        Nonnegative numbers that sum to 1

    Attributes
    ----------
    q : see Parameters
    Q : array_like(float)
        The cumulative sum of q

    """

    def __init__(self, q):
        self._q = q
        self.Q = cumsum(q)

    @property
    def q(self):
        """
        Getter method for q.

        """
        return self._q

    @q.setter
    def q(self, val):
        """
        Setter method for q.

        """
        self._q = val
        self.Q = cumsum(val)

    def draw(self, k=1):
        """
        Returns k draws from q.

        For each such draw, the value i is returned with probability
        q[i].

        Parameters
        -----------
        k : scalar(int), optional
            Number of draws to be returned

        Returns
        -------
        array_like(int)
            An array of k independent draws from q

        """
        return self.Q.searchsorted(uniform(0, 1, size=k))
    
def mc_sample_path(P, init=0, sample_size=1000):
    """
    Markov Chain Path
    
    """
    # === set up array to store output === #
    X = np.empty(sample_size, dtype=int)
    if isinstance(init, int):
        X[0] = init
    else:
        X[0] = DiscreteRV(init).draw()

    # === turn each row into a distribution === #
    # In particular, let P_dist[i] be the distribution corresponding to the
    # i-th row P[i,:]
    n = len(P)
    P_dist = [DiscreteRV(P[i, :]) for i in range(n)]

    # === generate the sample path === #
    for t in range(sample_size - 1):
        X[t+1] = P_dist[X[t]].draw()

    return X

def compute_fixed_point(T, v, error_tol=1e-3, max_iter=50, verbose=1, *args,
                        **kwargs):
    """
    Computes and returns :math:`T^k v`, an approximate fixed point.

    Here T is an operator, v is an initial condition and k is the number
    of iterates. Provided that T is a contraction mapping or similar,
    :math:`T^k v` will be an approximation to the fixed point.

    Parameters
    ----------
    T : callable
        A callable object (e.g., function) that acts on v
    v : object
        An object such that T(v) is defined
    error_tol : scalar(float), optional(default=1e-3)
        Error tolerance
    max_iter : scalar(int), optional(default=50)
        Maximum number of iterations
    verbose : bool, optional(default=True)
        If True then print current error at each iterate.
    args, kwargs :
        Other arguments and keyword arguments that are passed directly
        to  the function T each time it is called

    Returns
    -------
    v : object
        The approximate fixed point

    """
    iterate = 0
    error = error_tol + 1
    while iterate < max_iter and error > error_tol:
        new_v = T(v, *args, **kwargs)
        iterate += 1
        error = np.max(np.abs(new_v - v))
        if verbose:
            print("Computed iterate %d with error %f" % (iterate, error))
        try:
            v[:] = new_v
        except TypeError:
            v = new_v
    return v
```

```python
import matplotlib.pyplot as plt

def compute_asset_series(ip, T=500000, verbose=False):
    """
    Given optimal policies, simulate assets after time T.
    
    """
    # Shorten and define variables
    Pi, z_vals, R = ip.Pi, ip.z_vals, ip.R
    
    v_init = ip.initialize()
    v = compute_fixed_point(ip.bellman_operator, v_init, verbose=True)
    
    c = ip.bellman_operator(v, return_policy=True)
    cf = lambda a, i_z: interp(a, ip.asset_grid, c[:, i_z])
    
    a = np.zeros(T+1)
    z_seq = mc_sample_path(Pi, sample_size=T)
    for t in range(T):
        i_z = z_seq[t]
        a[t+1] = R * a[t] + z_vals[i_z] - cf(a[t], i_z)
    return a

ip = IndividualProblem(r=0.03, grid_max=4)
a = compute_asset_series(ip, T=10000)

# Display graph of asset distribution
%matplotlib inline
fig, ax = plt.subplots(figsize=(10, 8))
ax.hist(a, bins=20, alpha=0.5, normed=True)
ax.set_xlabel('assets')
ax.set_xlim(-0.05, 0.75)
plt.show()

print a
```
